```python
#21bds0384
import pandas as pd
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/TravelMode.csv"
df = pd.read_csv(url)


df.info()
```

```python
df.head()
```

```python
# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()
```

```python

new_row = {'rownames': 'utsavpoudel', 'individual': '21bds0384', 'mode': None, 'choice': None,
           'wait': None, 'vcost': None, 'travel': None, 'gcost': None, 'income': None, 'size': None}


df = pd.concat([pd.DataFrame([new_row]), df], ignore_index=True)


df.head()
```

Central Tendency for wait Column

```python
mean_wait = df['wait'].mean()
median_wait = df['wait'].median()
mode_wait = df['wait'].mode()[0]  # Mode can return multiple values, take the first

print(f"Mean of wait: {mean_wait}")
print(f"Median of wait: {median_wait}")
print(f"Mode of wait: {mode_wait}")
```

Measure of Dispersion for wait Column

```python
std_wait = df['wait'].std()
var_wait = df['wait'].var()

print(f"Standard Deviation of wait: {std_wait}")
print(f"Variance of wait: {var_wait}")
```

Quartile Ranges and IQR for wait Column

```python
q1_wait = df['wait'].quantile(0.25)
q3_wait = df['wait'].quantile(0.75)
iqr_wait = q3_wait - q1_wait

print(f"Q1 (25th percentile) of wait: {q1_wait}")
print(f"Q3 (75th percentile) of wait: {q3_wait}")
print(f"IQR of wait: {iqr_wait}")
```

 Correlation Between vcost and gcost

```python
correlation_vcost_gcost = df['vcost'].corr(df['gcost'])
print(f"Correlation between vcost and gcost: {correlation_vcost_gcost}")
```
