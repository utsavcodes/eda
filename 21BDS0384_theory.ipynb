# kmeans clustering elbow method, heirarecial clustering using all distance method taught on class, PCA and followed by regression model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import cdist
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/TravelMode.csv"
df = pd.read_csv(url)

print("Dataset Head:\n", df.head())

df_selected = df[['vcost', 'travel']]
print("\nSelected Columns (vcost, travel):\n", df_selected.head())

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_selected)

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('21bds0384 - Elbow Method For Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(df_scaled)

plt.scatter(df_scaled[y_kmeans == 0, 0], df_scaled[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(df_scaled[y_kmeans == 1, 0], df_scaled[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(df_scaled[y_kmeans == 2, 0], df_scaled[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(df_scaled[y_kmeans == 3, 0], df_scaled[y_kmeans == 3, 1], s=100, c='purple', label='Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.title('21bds0384 - Clusters of Customers (K-Means)')
plt.xlabel('Vcost')
plt.ylabel('Travel')
plt.legend()
plt.show()

distance_methods = ['euclidean', 'cityblock', 'chebyshev', 'canberra', 'hamming', 'minkowski']

def plot_dendrogram(distance_method):
    distance_matrix = cdist(df_scaled, df_scaled, metric=distance_method)
    Z = linkage(distance_matrix, method='ward')
    plt.figure(figsize=(10, 7))
    dendrogram(Z)
    plt.title(f'21bds0384 - Dendrogram ({distance_method.capitalize()} Distance)')
    plt.xlabel('Samples')
    plt.ylabel('Distance')
    plt.show()

for method in distance_methods:
    plot_dendrogram(method)

pca = PCA()
df_pca = pca.fit_transform(df_scaled)

print("\nExplained variance ratio (Eigenvalues):\n", pca.explained_variance_ratio_)
print("\nPrincipal Components Loadings:\n", pca.components_)

plt.figure(figsize=(10, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=y_kmeans, cmap='viridis')
plt.title('21bds0384 - PCA - First two components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.title('21bds0384 - Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()

n_comp = 2
pca_selected = PCA(n_components=n_comp)
df_pca_selected = pca_selected.fit_transform(df_scaled)

plt.figure(figsize=(10, 6))
plt.scatter(df_pca_selected[:, 0], df_pca_selected[:, 1], c=y_kmeans, cmap='viridis')
plt.title(f'21bds0384 - PCA with {n_comp} Components')
plt.xlabel(f'Principal Component 1')
plt.ylabel(f'Principal Component 2')
plt.show()
```

# regression model on this dataset along with the metrices and acuracy and classification results

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score


url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/TravelMode.csv"
data = pd.read_csv(url)


print(data.head())


data['choice'] = data['choice'].map({'yes': 1, 'no': 0})


data = data.drop(columns=['rownames', 'individual'])


data = pd.get_dummies(data, columns=['mode'], drop_first=True)


X = data.drop(columns=['choice'])
y = data['choice']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
